# Analyse der Unfälle in DE

# Datensatz inizialisieren

```python
import pandas as pd

# Datei einlesen
df = pd.read_csv('csv/Unfallorte2023_LinRef.csv', delimiter=';', dtype={'UIDENTSTLAE': str})


# Die ersten Zeilen anzeigen
print(df.head())
```

```python
print(df.info())
```

```python
df['XGCSWGS84'] = df['XGCSWGS84'].str.replace(',', '.').astype(float)
df['YGCSWGS84'] = df['YGCSWGS84'].str.replace(',', '.').astype(float)
```

```python
# Anzahl der Unfälle pro Jahr
print(df['UJAHR'].value_counts())

# Anzahl der Unfälle pro Monat
print(df['UMONAT'].value_counts())
```

## Auswertung der Unfallzahlen nach Monaten

```python
print(df['UMONAT'].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt

df['UMONAT'].value_counts().sort_index().plot(kind='bar', figsize=(10, 6))
plt.title('Unfälle nach Monat im Jahr 2023')
plt.xlabel('Monat')
plt.ylabel('Anzahl der Unfälle')
plt.xticks(range(12), ['Jan', 'Feb', 'Mär', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dez'], rotation=45)
plt.show()
```

```python
# Manuelle Zuordnung der Monatszahlen zu Monatsnamen
monatsnamen = {
    1: "Jan", 2: "Feb", 3: "Mär", 4: "Apr",
    5: "Mai", 6: "Jun", 7: "Jul", 8: "Aug",
    9: "Sep", 10: "Okt", 11: "Nov", 12: "Dez"
}

    #for monat in monatsnamen:
     #   print(monatsnamen[monat])

# Schritt 1: Häufigkeiten der Unfälle in jedem Monat berechnen
monats_anzahl = df['UMONAT'].value_counts()

# Schritt 2: Werte nach Anzahl sortieren
sortierte_monats_anzahl = monats_anzahl.sort_values()

    #print(sortierte_monats_anzahl)

# Schritt 3: Monatsnamen zuordnen
sortierte_monatsnamen = []

for monat in sortierte_monats_anzahl.index:
    sortierte_monatsnamen.append(monatsnamen[monat])

    
# Schritt 4: Erstellen eines neuen DataFrame für den Plot
sortierte_daten = pd.DataFrame({
    'Monat': sortierte_monatsnamen,
    'Unfälle': sortierte_monats_anzahl.values
})

# Schritt 5: Plotten der Werte
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(sortierte_daten['Monat'], sortierte_daten['Unfälle'])
plt.title("Unfälle pro Monat (aufsteigend sortiert)")
plt.xlabel("Monat")
plt.ylabel("Anzahl der Unfälle")
plt.grid(axis='y', alpha=0.7)


plt.savefig("BilderPlots/UnfälleProMonat(aufsteigend_sortiert).png")
plt.show()
plt.close()
```

## Die Analyse der Unfallzahlen zeigt folgende wichtige Erkenntnisse:

- Monat mit den wenigsten Unfällen:

    Der Februar verzeichnete die wenigsten Unfälle mit 15.646 Vorfällen. Dies könnte auf die geringere Anzahl an Tagen im Monat und eventuell weniger Verkehr durch winterliche Bedingungen zurückzuführen sein.

- Monat mit den meisten Unfällen:

    Der Juni ist der unfallträchtigste Monat mit 29.703 Unfällen. Eine mögliche Erklärung könnte die höhere Verkehrsdichte während der Sommermonate sein, in denen viele Menschen reisen und die Straßen stärker frequentiert werden.

- Saisonale Schwankungen:

    Die Sommermonate (Juni bis September) weisen im Vergleich zu den Wintermonaten tendenziell höhere Unfallzahlen auf. Dies deutet darauf hin, dass das Verkehrsaufkommen und die Fahrbedingungen in den wärmeren Monaten eine größere Rolle bei der Unfallhäufigkeit spielen.

- Einfluss der Wetterbedingungen:

    Die Wintermonate (Dezember, Januar, Februar) liegen in der unteren Hälfte der Unfallstatistik, was auf mögliche Wetterbedingungen (z. B. Schnee und Eis) und vorsichtigeres Fahrverhalten zurückzuführen sein könnte.

- Besondere Beobachtungen:

    Der Mai und der Juli weisen ähnlich hohe Unfallzahlen auf (jeweils über 26.000), was darauf hinweist, dass diese Monate auch stark vom Verkehrsaufkommen beeinflusst werden.

- Auffällig ist, dass der Oktober mit 23.984 Unfällen weniger stark betroffen ist, obwohl es ein Übergangsmonat mit möglichen erhöhten Verkehrsrisiken (z. B. kürzere Tage, Regen) ist.

# Unfälle wochen Tage

```python
print(df['UWOCHENTAG'].value_counts().sort_index())

wochenTage = {
    1: "Mo", 2: "Di", 3: "Mi", 4: "Do", 5:"Fr", 6: "Sa", 7: "So"
}

df['UWOCHENTAG'].value_counts().sort_index().plot(kind='bar', figsize=(10, 6))
plt.title('Unfälle nach Wochentagen im Jahr 2023')
plt.xlabel('WochenTage')
plt.ylabel('Anzahl der Unfälle')
plt.xticks(range(7), [wochenTage[i+1] for i in range(7)], rotation=0)
plt.grid(axis='y', alpha=0.7,color='gray')


plt.savefig("BilderPlots/unfälleNachWochenTagen.png")
plt.show()
plt.close()
```

## Beobachtungen & Interpretationen:

- Höchste Unfallzahlen:

    Der Samstag weist mit 43.863 Unfällen die höchste Zahl auf.
    Dies könnte auf erhöhtes Verkehrsaufkommen durch Freizeit- und Ausflugsaktivitäten zurückzuführen sein.

- Zweite Spitze:

Der Freitag folgt knapp mit 43.144 Unfällen. Dies liegt vermutlich daran, dass viele Menschen am Freitagnachmittag oder -abend ins Wochenende starten, wodurch es zu mehr Verkehr und Unfällen kommt.

- Niedrigste Unfallzahlen:

Der Montag hat mit 26.021 Unfällen die geringsten Zahlen. Dies deutet darauf hin, dass der Verkehr zu Wochenbeginn möglicherweise geordneter abläuft oder weniger Freizeitverkehr vorhanden ist.

- Wochenmitte:

Dienstag, Mittwoch und Donnerstag weisen ähnlich hohe Unfallzahlen auf, jeweils zwischen 40.872 und 41.452 Unfällen. Hier spiegeln sich typische Werktage wider, mit regelmäßiger Pendleraktivität und Berufsverkehr.

- Sonntag:

Der Sonntag hat 32.683 Unfälle, was deutlich weniger als Samstag, aber mehr als Montag ist. Möglicherweise liegt dies daran, dass viele Menschen sonntags weniger unterwegs sind oder Freizeitaktivitäten anders verteilt sind.


- Interpretationen
  
    - Freizeit- und Pendlerverkehr: Die hohen Unfallzahlen an Samstagen und Freitagen könnten auf eine Mischung aus Freizeit- und Berufsverkehr zurückzuführen sein.
    - Verkehrsvolumen: Generell lässt sich sagen, dass das Verkehrsvolumen und das Unfallrisiko unter der Woche höher ist als am Montag oder Sonntag.
    - Saisonale Effekte: In den Daten ist nicht berücksichtigt, ob Feiertage oder besondere Ereignisse das Unfallgeschehen beeinflusst haben.
